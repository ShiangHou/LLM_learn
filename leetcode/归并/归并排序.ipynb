```python
#归并排序就是说，主函数是一个递归的过程，然后让左有序，右有序，左右合并，就行

#这个函数的意思是说，让一个数组nums在l到r上有序
def mergesort(nums:list,l:int,r:int)->list:
    if l == r:
        return 0
    m = (l+r) //2
    mergesort(nums,l,m)
    mergesort(nums,m+1,r)
    merge(nums,l,m,r)
    return nums





```

```python
#下面这个是合并过程，即在左右有序的情况下合并
#做法是，先定义一个虚拟的，然后两个指针在两端刷，不断把小的打入新的，最后刷回数组
def merge(nums,l,m,r):
    temp = []
    i,j = l,m+1
    while i <=m and j<=r:
        if nums[i]<= nums[j]:
            temp.append(nums[i])
            i += 1
        else:
            temp.append(nums[j])
            j += 1
    while i <= m:
        temp.append(nums[i])
        i += 1
    while j <=r:
        temp.append(nums[j])
        j += 1
    #最后刷回去
    nums[l:r+1] = temp
    


```

```python
nums = [2,5,8,3,12,7,9,2,0]
print(mergesort(nums,0,len(nums)-1))
```
