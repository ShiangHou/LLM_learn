```python
# def decodeString(s: str) -> str:
    #思路是，除了]都全部压入
    #遇到]后，到栈里面取，取到第一个次遇到[和前面的数字，计算下来，然后再压回栈
stack = []
s = "3[a]2[bc]"
for k in s:

    if k.isdigit() or k.isalpha or k == "[":
        stack.append(k)
    elif k == "]":
        temp_alpha = ""#准备乘的字母
        while stack[-1] != "[":
            temp_alpha += stack.pop()
        #遇到了[，弹出[
        stack.pop()
        #找前面的数字
        temp_num = int(stack.pop())
        temp = ""
        for _ in range(temp_num):
            temp += temp_alpha
        #最后再压入栈
        stack.append(temp)
print("".join(stack))#最后输出合并的栈里面的所有内容
```
