首先是栈实现队列

思路是这样，首先准备两个栈，一个叫in，一个叫out

当加入元素时，一个一个加入到in栈

当需要弹出元素时 
    1）判读一下out栈是不是空的，如果不是空的，直接从out栈弹出就行
    2）如果out栈是空的，那么就把in栈里的全部元素一个一个弹出，然后压入out栈中，最后弹出out栈顶


用队列来去实现栈

当加入元素的时候，我们需要记录一个数，这个数是压入前这个队列中元素的个数size

当加入元素时，
    1）加入元素到队列中
    2）将队列中一个元素弹出，然后压入队列，这个过程进行size次（意味着之前全部的元素的次序颠倒并且到了屁股后面）
    3）记得把size++

当弹出元素时，正常弹出队列的元素就行
