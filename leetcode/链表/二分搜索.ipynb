二分搜索 就是说在一个有序数组中，要找到某一个数

```python
#在一个有序数组中，找一个数
def exist(nums:list,x:int):
    if nums == None or len(nums) == 0:
        return False
    l,r = 0, len(nums)-1
    while l <=r:
        m = (l+r) //2
        if nums[m] == x:
            return True
        else:
            if nums[m] > x:
                r = m-1
            else:
                l = m+1
    return False
```

```python
#那么还有一些问题，依旧是在一个有序数组中，比如找到小于等于某个数的位置
#比如，找大于等于x的最左位置（最右位置没有意义，因为只需要看最右边的就行）
def findleft(nums:list,x:int):
    if nums == None or len(nums) == 0:
        return -1
    ans = -1
    l,r = 0,len(nums)-1
    while l <= r:
        m = (l+r) //2
        if nums[m]>=x:#那么所有的右边都是大于x的，王左边缩小范围，同时记录一下这个答案
            ans = m
            r = m-1
        else:
            l = m+1
    return ans
```

```python
nums = [4,6,8,12,17,20]
x = 10
print(findleft(nums,x))
```

```python
def findright(nums:list, x:int):
    if nums == None or len(nums) == 0:
        return -1
    l,r  = 0,len(nums)-1
    while l <= r:
        m = (l+r) //2
        if nums[m]<=x:#那么所有的左边都是小于x的，往右边缩小范围，同时记录一下这个答案
            ans = m
            l = m+1
        else:
            r = m-1
    return ans
```

```python
nums = [4,6,8,12,17,20]
x = 10
print(findright(nums,x))
```

二分搜索不一定发生在有序数组上，比如峰值问题。给定一个数组，任意一个峰值，其中峰值指的是 i-1 i i+1 如果nums[i]位置上比剩下两个都大，那么位置i就是峰值

首先，数组的两头，左边是上扬的，右边是下降的，那么中间必有一个峰值，取mid看一下，如果mid符合条件，返回就行

如果mid是上扬的，那么说明，mid+1 n上必有峰值，那么缩小继续二分就行，同理，如果mid是下降的，那么0-mid-1上必有峰值，往左二分就行

```python
def findmax(nums:list):
    if nums == None or len(nums) == 0:
        return -1
    #这里要先验证一些特殊情况

    #如果数组的长度是1，直接返回
    if len(nums) == 1:
        return 0
    #如果数组的长度大于等于2，要看一下两头是不是，因为假设了 -1 位置是无穷小，n+1位置是无穷小
    if len(nums) >= 2:
        if nums[0] > nums[1]:#验证0位置，如果1位置是小的，那么0位置就是极大值
            return 0
        if nums[len(nums)-1] > nums[len(nums)-2]:
            return len(nums)-1

    #0和n都不是了，所以要从1和n-2开始，要不然会出错
    l,r = 1,len(nums)-2 
    while l <= r:
        m = (l+r)//2
        if nums[m-1] > nums[m]:
            r = m-1
        else:
            if nums[m] < nums[m+1]:
                l = m+1
            else:
                return m
    return -1
```
