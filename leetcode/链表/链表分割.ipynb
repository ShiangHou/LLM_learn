```python
#题目就是，给你一串链表，需要将他们按照某个划分，但保持原来的相对位置不变
#比如链表是 3，7 9 6 4 8，把小于等于4的放左边，大于等于4的放右边
#那么划分后就是 3，4，7，9，6，8
```

```python
class ListNode:
    def __init__(self,value = 0,next = None):
        self.value = value
        self.next = next


head7 = ListNode(5,None)
head6 = ListNode(9,head7)
head5 = ListNode(1,head6)
head4 = ListNode(4,head5)
head3 = ListNode(8,head4)
head2 = ListNode(7,head3)
head = ListNode(2,head2)
#2，7，8，4，1，9，5
# 按4排序应该是
#2，1，4，7，8，9，5
```

思路是，需要四个变量就可以，一个是小头和小尾巴，一个是大头和大尾巴
开始遍历，遇到小的接在小尾后面，然后小尾巴往后走，切换掉None，遇到大的也是，最后让小尾链接大头就行

```python
def mergeList(head:ListNode,x:int)->ListNode:
    l_head = None
    l_tail = None
    r_head = None
    r_tail = None
    cur = head
    while cur != None:
        temp = cur.next#先存起来他的下一个
        cur.next = None#再把当下的和后面的断掉


        if cur.value <= x:#当走到一个小于的时候
            if l_head == None:#如果说此时还没建立，就把他们都加上
                l_head = cur
            else:#之前有了，就把tial处理一下
                l_tail.next = cur
            #最后l_tail肯定是移动到当前的真正的“最小尾巴”
            l_tail = cur
        else:
            if r_head == None:
                r_head  = cur
            else:
                r_tail.next = cur
                
            r_tail = cur
        #把刚才存的再粘贴回去
        cur = temp
    #最后链接一下头尾
    if l_tail: # 如果左边有节点
        l_tail.next = r_head
        return l_head
    else: # 如果左边没节点（l_head也是None），直接返回右边
        return r_head


```
